# Read data address

```python
import os

list_folder = os.listdir('/content/Persian-poems/database')
print(len(list_folder))
list_path = []
for i in list_folder:
  path = '/content/Persian-poems/database'+'/'+i
  list_path.append(path)
```

# import libraries

```python
from haraai_clean import General_normalization
from normalizer_hazm import clean_normal_data_hazm, clean_tags, correct_space
import check
```

# Save data function

```python
def save_data(name, data):
  fp = open(name, 'w')
  fp.writelines(data)
  fp.close()
```

# dataset info

```python
fp = open(list_path[0])
texts = fp.readlines()
print("some example: ", texts[140:145])
print("check english char: ", check.check_english_char(texts[0:20]))
print("document lines are ", len(texts))
```

```python
general = General_normalization()

tags = ['<[SOFM]>' ,'<[SOT]>', '<[EOSM]>', '<[SOFM]>',
        '<[EOFMMBT]>','<[SOSMMBT]>','<[EOFM]>', '<[SOSM]>', '<[SOSM]>',
        '<[EOT]>']

list_clean = []
test_text = texts[0:1500]
for i in test_text:
  data = clean_tags(tags, [i])[0]  # حذف همه ی تگ ها
  data = correct_space(data) # # اصلاح فاصله گذاری

  if len(data)>3:
    data = clean_normal_data_hazm(data) # remove_specials_chars, persian_style, unicodes_replacement, seperate_mi

    data = general.arabic_correction(data)
    data = general.semi_space_correction(data)
    data = general.space_between_punctuations_and_text(data)
    data = general.special_alphabet_correction(data)
    data = general.alphabet_correction(data)
    data = data.replace('\u200c', ' ') # حذف نیم فاصله

    output = general.remove_earab(data)
    output = output.strip() # حذف فاصله ها از اول و آخر
    list_clean.append(output)

print("check english char: ", check.check_english_char(list_clean))
print("check number: ", check.check_numbers(list_clean))
print("check special char: ", check.check_special_charecter(list_clean)) # except ،
print(list_clean)
```
